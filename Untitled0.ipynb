Types of Data (Simple Classification)
1. Numerical Data (Quantitative)

Data that is measured or counted and expressed in numbers.

Types:

Discrete Data

Countable values

No fractions

Examples: number of students, number of games, rank

Continuous Data

Measurable values

Can include decimals

Examples: height, weight, temperature, sales

2. Categorical Data (Qualitative)

Data that describes qualities or categories, not numbers.

Types:

Nominal Data

No order or ranking

Examples: color, gender, blood group, platform, genre

Ordinal Data

Has order or ranking

Differences not measurable

Examples: grades (A, B, C), education level, ratings (good, better, best)

**when we want to show propotion of some thing we use pie chart **


*In mathematics and statistics, a proportion is a part, share, or number considered in relation to a whole. It's often expressed as a fraction, decimal, or percentage*

```python
sns.countplot( x ='Genre',data = Data,order=Data['Genre'].value_counts().index,color='orange',width = 0.2)
plt.xticks(rotation = 90,fontsize = 10)
plt.show()
#we can plot this graph using matplotlib but using seaborn we have to  write less code
```

```python
plt.pie(y_lable,labels = x_lable,autopct = '%1.2f%%')
plt.title('Various Genre Contributions',fontsize = 20)
plt.show()
```

```python
sales_data = Data[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
```

```python
sales_data
```

```python
region_sales = sales_data.T.sum(axis ='columns')
# T is used to transpose columns into rows and sum the values across columns.
```

```python
region_sales
```

```python
plt.pie(region_sales,labels=region_sales.index,autopct = '%1.2f%%', startangle= 90,explode=(0.2,0,0,0))
plt.title('Total Sales across various region', fontsize = 20)
```
